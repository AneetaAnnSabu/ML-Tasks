### <font color='red'> 1. Logistic Regression <a id ='model_1'></a>

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

#### <font color =' Green'> Evaluation - Logistic Regression

```python
y_predict=lr.predict(x_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
cm
```

```python
#heatmap
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
#### In the confusion matrix ,diagonal values are correct predictions.others are wrong predictions here only 1 value is predicted wrongly.
```

```python
from sklearn.metrics import classification_report
print('Classification report:')
print(classification_report(y_predict,y_test))
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_test,y_predict)
print(f'Accuracy:{accuracy}')
```

### This model has an accuracy of ~97%.
##### This high accuracy suggests an ideal model, performing with exceptional precision and recall.
##### The precision value, also near 1, signifies that most of the model's positive predictions are correct. 
##### The recall value is also close to 1, demonstrating that the model successfully identifies nearly all positive cases.
##### f1 score is also close to 1 indicates that the model maintains an excellent equilibrium between precision and recall, ensuring comprehensive and accurate predictions.

### <font color='red'>1.K-MEANS CLUSTERING <a id='model_5'></a>

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
from sklearn.cluster import KMeans 

kmeans = KMeans(n_clusters = 3, random_state = 42)
```

```python
kmeans.fit(df[iris.feature_names])
```

```python
predicted_test = kmeans.predict(df[iris.feature_names])
predicted_test
```

```python
iris.target_names
```

```python
from sklearn.cluster import KMeans
X=df.iloc[:,[0,1,2,3]].values

k_range=range(1,10)
sse=[] #sum of squared error

for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(X)
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel("K value")
plt.ylabel("SSE")
plt.plot(k_range, sse)
```

```python
#Based on the graph, the "elbow" appears to be at 3 clusters, where the rate of decrease in WCSS starts to slow down.
```

```python
#Training the model with k=3
km = KMeans(n_clusters=3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_predicted = km.fit_predict(X)
y_predicted
```

```python
km.cluster_centers_
```

```python
#centroid of the model
plt.scatter(X[y_predicted == 0, 2], X[y_predicted == 0, 3], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_predicted == 1, 2], X[y_predicted == 1, 3], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_predicted == 2, 2], X[y_predicted == 2, 3], s = 100, c = 'green', label = 'Cluster 3')

plt.scatter(km.cluster_centers_[:, 2], km.cluster_centers_[:,3], s = 400, marker="*" ,c = 'black', label = 'Centroids')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.legend()
```

### <font color='green'> Evaluation- KMeans Clustering

```python
from sklearn.metrics import silhouette_score

X=df.iloc[:,[0,1,2,3]].values

silhouette_scores = []

# Try k values from 2 to 10
k_values = range(2, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)
silhouette_scores
```

```python
plt.figure(figsize=(8, 5))
plt.plot(k_values, silhouette_scores)
plt.title('Silhouette Score v/s Number of Clusters (k)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.show()
```

#### Highest value of silhouette score is obtained when k=2.

```python
#training the model with k=2
km2 = KMeans(n_clusters=2,random_state = 25)
y_predicted6 = km2.fit_predict(x)
y_predicted6
```

```python
#Checking centroid of the model
km2.cluster_centers_
```

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[y_predicted6 == 0, 0], x[y_predicted6 == 0, 1], s = 100, c = 'red', label = 'cluster 1')
plt.scatter(x[y_predicted6 == 1, 0], x[y_predicted6 == 1, 1], s = 100, c = 'blue', label = 'cluster 2')

# Plotting the centroids of the clusters
plt.scatter(km2.cluster_centers_[:, 0], km2.cluster_centers_[:,1], s = 400, marker="*" ,c = 'black', label = 'Centroids')

plt.legend()
```
